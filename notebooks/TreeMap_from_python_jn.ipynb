###### You can use R in python notebook with the R-magic. After you have loaded the extension shown above, just put '%%R' on top of the cells you want to use R in. 

```python
%%R
library('treemap') #load treemap, make sure its installed in your env
tree_df = read.csv('./csv/tree_df.csv', row.names='X') #load the data
treemap(tree_df, index=c('imod_type', 'imod_name'), vSize='gene_count') #make the treemap
```

#### If you want to save the figure as svg add the following lines.

```python
%%R
library('treemap')

svg(filename="./treemap.svg", #file where it will be saved
    width=4, #svg params, change as you need
    height=6, 
    pointsize=12)
tree_df = read.csv('./csv/tree_df.csv', row.names='X') #load data
treemap(tree_df, index=c('imod_type', 'imod_name'), vSize='gene_count') #make treemap
dev.off() #close svg
```
